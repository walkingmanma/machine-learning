```python
# 计算每一天的自行车流量，将结果放到一个新的DateFrame中
#Convenience method for frequency conversion and resampling of timeseries.  Object must have a datetime-like index (DatetimeIndex,
#PeriodIndex, or TimedeltaIndex), or pass datetime-like values to the on or level keyword.
```

```python
daily = counts.resample('d',how='sum')
daily.head()
```

```python
daily['Total'] = daily.sum(axis=1)
```

```python
daily.head()
```

```python
daily = daily[['Total']]
```

```python
daily.head()
daily.index.dayofweek
```

```python
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
```

```python
for i in range(7):
    daily[days[i]] = (daily.index.dayofweek == i).astype(float)
```

```python
daily.head()
```

```python
# 我们在上述DataFrame 中添加上节假日的关系
from pandas.tseries.holiday import USFederalHolidayCalendar
```

```python
cal = USFederalHolidayCalendar()
```

```python
holidays = cal.holidays('2012','2016')
holidays
```

```python
holidaysSeries = pd.Series(1,index=holidays,name='holiday')
holidaysDataFrame = pd.DataFrame(holidaysSeries)
```

```python
daily = daily.join(holidaysDataFrame,join)
```
