```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# %config InlineBackend.figure_format = 'svg'
```

## Step 1 : get Feature and data Sample

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
x_sample = iris.data
y_sample = iris.target
feature_name = iris.feature_names
target_name = iris.target_names
```

## Step 2 : import sklean Decision Trees Classification and fit 

###### 1，DecisionTreesClassifier的参数  ？？？
##### 2，DecisionTreesClassifier optimal rule   ？？？
       gini 基尼值最大，或者说是Gini_index基尼指数最小作为划分属性
       
##### 3，DecisionTreesClassifier 剪枝策略


```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model = model.fit(x_sample,y_sample)
feature_name
```

```python
target_name
```

## Step 3 : until now , we have finish the trainng and got a tree , so we have a look at the branch

```python
import graphviz
from sklearn import tree
dot_data = tree.export_graphviz(model, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True) 
graph = graphviz.Source(dot_data) 
   
```
